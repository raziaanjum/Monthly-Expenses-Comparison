```python
### importing python libraries ######## 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading and loading dataset from an Excel file with file path 
expense = pd.read_excel("C:/Users/FIJESTU/Desktop/MonthlyExpensesDataset.xlsx")
print(expense)
```

```python
# Creating a pandas DataFrame named `mon_exp` using the dataset from `expense`
mon_exp = pd.DataFrame(expense)
print(mon_exp)
```

```python

# Define the correct month order as a categorical type
mon_exp["Month"] = pd.Categorical(
    mon_exp["Month"], categories=["January", "February", "March", "April", "May", "June"],
    ordered=True )

# Group by 'Month' and sum values for multiple columns
grouped_by = mon_exp.groupby("Month", observed=False )[["Transportation", "Food", "Education"]].sum()


# Plot each column as a separate bar in the bar plot
colors = ["green", "orange", "grey", "brown", "blue"]
grouped_by.plot(width = 0.8,kind="bar", color=colors)

# X-axis, Y-axis labels and positioning with Title 
plt.xlabel('Months')
plt.ylabel('Amount')
plt.title('Monthly Expenses Comparison')

# Adjusting the x-axis labels to be horizontal
plt.xticks(rotation=0)

# Adding the legend and displaying the plot
plt.legend()
plt.tight_layout()
plt.show()
```
